```python
#CSV file only containe the number of Customers
#this code automatically detect which number is on whatsapp or not, and send the message

import csv
import itertools
file_name = "filename.csv"
try:
    csvfile = open(file_name, "rt", encoding="utf8")
except:
    print("File not found")
csvReader = csv.reader(csvfile, delimiter=",")
l =  list()
for row in csvReader:
    l.append((row[:]))
print(l)
l  = list(set(list(itertools.chain.from_iterable(l)))) #Line Credit Data Analytics Ireland https://dataanalyticsireland.ie/category/unhashable-type-error/
```

```python
#Version4
from selenium import webdriver
import win32api, win32con
import keyboard
import click
import time
successfully_send = 0
failed_List = []
failtosend =0
driver = webdriver.Chrome(executable_path=".\chromedriver.exe")
driver.implicitly_wait(10)
driver.get("https://web.whatsapp.com/")
driver.implicitly_wait(10)
time.sleep(40)
all_names = ['Muhammad Azhar', 'Muhammad Azhar','Muhammad Azhar','Muhammad Azhar', 'Muhammad Azhar','Muhammad Azhar']
msg = 'Auto Test SeliniumlL'
count = 1


for name in all_names:
    try:
        time.sleep(3)
        driver.implicitly_wait(10)
        user = driver.find_element_by_xpath('//*[@id="side"]/header/div[2]/div/span/div[2]/div/span')
        user.click()
        time.sleep(2)
        user = driver.find_element_by_xpath('//*[@id="app"]/div[1]/div[1]/div[2]/div[1]/span/div[1]/span/div[1]/div[1]/div/label/div/div[2]').send_keys(name)
        
        driver.implicitly_wait(10)
        time.sleep(5)
        user= driver.find_element_by_xpath('//*[@id="app"]/div[1]/div[1]/div[2]/div[1]/span/div[1]/span/div[1]/div[2]/div[1]/div/div/div[2]/div/div/div[2]').click()
        #//*[@id="app"]/div[1]/div[1]/div[2]/div[1]/span/div[1]/span/div[1]/div[2]/div[1]/div/div/div[2]/div/div/div[2]
        #//*[@id="app"]/div[1]/div[1]/div[2]/div[1]/span/div[1]/span/div[1]/div[2]/div[1]/div/div/div[2]/div/div/div[2]

        #webbrowser.open(url, new=0, autoraise=True)


        #players = driver.find_elements_by_xpath('//td[@class="name"]')
        #user.click()
        driver.implicitly_wait(10)


        driver.implicitly_wait(10)
        msg_box = driver.find_element_by_xpath('//*[@id="main"]/footer/div[1]/div[2]/div/div[2]').send_keys(msg)
        driver.implicitly_wait(10)
        button = driver.find_element_by_xpath('//*[@id="main"]/footer/div[1]/div[3]/button/span')
        driver.implicitly_wait(10)
        button.click()
    
        user = driver.find_element_by_xpath('//*[@id="side"]/div[1]/div/label/div/div[2]').clear()
        driver.implicitly_wait(10)
        print("Successfully Send Message to : "+ name + '\n')
        successfully_send+=1
    except:
        print("Cannot find Target: " + name)
        failed_List.append(name)
        failtosend+=1
        
print("\n\n\nSuccessFully Send Messages To "+ str(successfully_send) + "Contacts")
print("\nFailed to Send Messages To "+ str(failtosend) + "Contacts")
print(failed_List)
```
